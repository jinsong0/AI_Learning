```python
import gradio as gr
import dashscope
from dashscope import ImageSynthesis
import os
import time

# ğŸ”‘ è®¾ç½® API Keyï¼ˆå»ºè®®ä½¿ç”¨ç¯å¢ƒå˜é‡ï¼‰
dashscope.api_key = os.getenv("DASHSCOPE_API_KEY", "sk-your-api-key-here")

def generate_image(prompt, size="1024*1024"):
    """è°ƒç”¨é€šä¹‰ä¸‡ç›¸æ–‡ç”Ÿå›¾"""
    try:
        response = ImageSynthesis.call(
            model="wan2.5-t2i-preview",
            prompt=prompt,
            size=size,
            n=1
        )
        if response.status_code == 200:
            image_url = response.output.results[0].url
            return image_url
        else:
            return f"âŒ å›¾åƒç”Ÿæˆå¤±è´¥: {response.code} - {response.message}"
    except Exception as e:
        return f"âš ï¸ é”™è¯¯: {str(e)}"

def generate_video(prompt):
    """è°ƒç”¨é€šä¹‰ä¸‡ç›¸æ–‡ç”Ÿè§†é¢‘ï¼ˆéœ€æƒé™ï¼‰"""
    try:
        response = ImageSynthesis.call(
            model="wan2.5-t2v-preview",
            prompt=prompt
        )
        if response.status_code == 200:
            video_url = response.output.results[0].url
            return video_url
        else:
            return f"âŒ è§†é¢‘ç”Ÿæˆå¤±è´¥: {response.code} - {response.message}"
    except Exception as e:
        return f"âš ï¸ é”™è¯¯: {str(e)}"

# Gradio ç•Œé¢
with gr.Blocks(title="æ¢¦æƒ³ç»˜å›¾æ¿ - æ–‡ç”Ÿå›¾ & æ–‡ç”Ÿè§†é¢‘") as demo:
    gr.Markdown("## âœ¨ æ¢¦æƒ³ç»˜å›¾æ¿ï¼šæ–‡å­—è½¬å›¾åƒä¸è§†é¢‘çš„é­”æ³•å·¥å‚")
    gr.Markdown("ç”±é˜¿é‡Œäº‘ DashScope é©±åŠ¨ Â· è¾“å…¥æ–‡å­—ï¼Œä¸€é”®ç”Ÿæˆé«˜æ¸…å›¾ç‰‡æˆ–çŸ­è§†é¢‘")

    with gr.Tab("ğŸ–¼ï¸ æ–‡ç”Ÿå›¾"):
        img_prompt = gr.Textbox(label="æè¿°ä½ æƒ³ç”Ÿæˆçš„å›¾åƒï¼ˆä¸­æ–‡/è‹±æ–‡ï¼‰", placeholder="ä¾‹å¦‚ï¼šä¸€åªç©¿ç€å®‡èˆªæœçš„çŒ«åœ¨æœˆçƒä¸Šæ¼«æ­¥ï¼Œèµ›åšæœ‹å…‹é£æ ¼")
        img_size = gr.Radio(choices=["1024*1024", "720*1280", "1280*720"], value="1024*1024", label="å›¾åƒå°ºå¯¸")
        img_btn = gr.Button("ç”Ÿæˆå›¾åƒ")
        img_output = gr.Image(label="ç”Ÿæˆç»“æœ", type="filepath")

        img_btn.click(
            fn=generate_image,
            inputs=[img_prompt, img_size],
            outputs=img_output
        )

    with gr.Tab("ğŸ¥ æ–‡ç”Ÿè§†é¢‘"):
        vid_prompt = gr.Textbox(label="æè¿°ä½ æƒ³ç”Ÿæˆçš„çŸ­è§†é¢‘ï¼ˆå»ºè®®ç®€æ´ï¼‰", placeholder="ä¾‹å¦‚ï¼šæµ·æµªæ‹æ‰“å²©çŸ³ï¼Œæ—¥è½æ—¶åˆ†ï¼Œé«˜æ¸…å†™å®")
        vid_btn = gr.Button("ç”Ÿæˆè§†é¢‘ï¼ˆéœ€å¼€é€šæƒé™ï¼‰")
        vid_output = gr.Video(label="ç”Ÿæˆç»“æœ")

        vid_btn.click(
            fn=generate_video,
            inputs=vid_prompt,
            outputs=vid_output
        )

    gr.Markdown("""
    > ğŸ’¡ æç¤ºï¼š
    > - æ–‡ç”Ÿå›¾å·²å¼€æ”¾ï¼Œæ–°ç”¨æˆ·æœ‰å…è´¹é¢åº¦ã€‚
    > - æ–‡ç”Ÿè§†é¢‘ç›®å‰ä¸ºé‚€æµ‹åŠŸèƒ½ï¼Œè¯·å‰å¾€ [DashScope æ§åˆ¶å°](https://dashscope.console.aliyun.com/) ç”³è¯·ã€‚
    > - ç”Ÿæˆç»“æœä¸ºå…¬ç½‘ URLï¼Œè‡ªåŠ¨ä¸‹è½½æ˜¾ç¤ºã€‚
    """)

if __name__ == "__main__":
    # å»ºè®®é€šè¿‡ç¯å¢ƒå˜é‡ä¼ å…¥ API Key
    # export DASHSCOPE_API_KEY="sk-xxxx"
    demo.launch()
```
