```python
import gradio as gr
```

```python
def echo(message):
    return message
```

```python
demo = gr.ChatInterface(
    fn = echo,
    title = "回声",
    description = "这是一个回声机器人"
)
```

```python
demo.launch(share=True)
```

```python
# step 1: 定义处理函数
def Hello(name):
    return f"你好，{name}，欢迎学习GRADIO!"
```

```python
# step 2&3 : 定义输入和输出
demo = gr.Interface(
    fn = Hello,        # 处理
    inputs = "text",   # 输入
    outputs = "text"   # 输出
)
```

```python
demo.launch()  # 启动应用
```
