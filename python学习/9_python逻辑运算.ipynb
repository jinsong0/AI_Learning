```python
# 逻辑运算符 and or not
# 逻辑运算符用于对条件进行判断，并返回布尔值（True或False）。
# and：如果左右两边的条件都为True，则返回True；否则返回False。
# or：如果左右两边的条件有一个为True，则返回True；否则返回False。
# not：对条件取反，如果条件为True，则返回False；如果条件为False，则返回True。
# 优先级：not > and > or
```

```python
house_work_count = int(input("请问你一共做了多少次家务？"))
read_envelopes_count = int(input("请问你一共读了多少封信？"))
shopping_count = int(input("请问你一共购买了多少件商品？"))
has_been_angry = input("你是否有过生气的经历？（是/否）")
if(house_work_count > 10 and read_envelopes_count > 1 and shopping_count > 4 and not has_been_angry):
    print("摩拳擦掌等待Switch的到来！")
else:
    print("继续保持沉默，继续努力！")
```
