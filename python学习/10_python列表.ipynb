# Python 列表（List）基本概念

## 什么是列表？

在 Python 中，**列表（`list`）** 是一种**可变的、有序的**数据结构，用于存储**一系列元素**。列表中的元素可以是任意类型（如整数、字符串、其他列表等），并且允许重复。

列表使用方括号 `[]` 定义，元素之间用逗号 `,` 分隔。

### 基本语法

```python
my_list = [1, 2, 3, "hello", True]

````python
shopping_list = ["键盘", "键帽"]
print(shopping_list)
## 添加元素到列表中 append()
shopping_list.append("显示器")
print(shopping_list)
````

````python
# 字符串是不可变的，所以不能对字符串的元素进行修改
s = "hello"
print(s.upper())
print(s)
````

````python
s = "hello"
s = s.upper()
print(s)
````

````python
## 列表是可变的，所以可以对列表的元素进行修改
shopping_list = ["键盘", "键帽"]
shopping_list.append("显示器")
print(shopping_list)
shopping_list = shopping_list.append("显示器")    # 错误写法
print(shopping_list)
````

````python
##删除列表中的元素 remove()
shopping_list = ["键盘", "键帽", "显示器"]
shopping_list.remove("显示器")
print(shopping_list)
````

````python
## python列表可以放不同类型的数据
list1 = ["hello"]
list1.append(66.6)
list1.append(True)
list1.append(None)
print(list1)
````

````python
## 列表和字符串一样都可以通过len()函数获取长度
shopping_list = ["键盘", "键帽", "显示器"]
len(shopping_list)
````

````python
## 通过索引访问列表中的元素
shopping_list = ["键盘", "键帽", "显示器"]
print(shopping_list[0])
print(shopping_list[1])
print(shopping_list[2])
print(shopping_list[len(shopping_list)-1])    # 也可以通过len()函数获取列表的长度再减1来访问最后一个元素
````

````python
## 修改列表中的元素
shopping_list = ["键盘", "键帽", "显示器"]
shopping_list[1] = "鼠标"
shopping_list
````

````python
## 针对列表的内置函数
num_list = [1, 13, -7, 2, 96]
print(max(num_list))  ## 打印列表里的最大值
print(min(num_list))  ## 打印列表里的最小值
print(sorted(num_list))  ## 打印排序后的列表

````
